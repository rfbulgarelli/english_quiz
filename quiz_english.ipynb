```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Lista completa de perguntas e respostas
questions = [
    {
        "question": "I always am happy when it’s sunny outside.",
        "options": [
            "I am always happy when it’s sunny outside.",
            "I always am happy when it’s sunny outside."
        ],
        "correct": "I am always happy when it’s sunny outside."
    },
    {
        "question": "She usually drinks coffee in the morning.",
        "options": [
            "She usually drinks coffee in the morning.",
            "She drinks usually coffee in the morning."
        ],
        "correct": "She usually drinks coffee in the morning."
    },
    {
        "question": "They often are going to the gym after work.",
        "options": [
            "They often go to the gym after work.",
            "They often are going to the gym after work."
        ],
        "correct": "They often go to the gym after work."
    },
    {
        "question": "Sometimes we eat out on weekends.",
        "options": [
            "Sometimes we eat out on weekends.",
            "We sometimes eat out on weekends."
        ],
        "correct": "Sometimes we eat out on weekends."
    },
    {
        "question": "We do not often visit our grandparents.",
        "options": [
            "We do not often visit our grandparents.",
            "We not often do visit our grandparents."
        ],
        "correct": "We do not often visit our grandparents."
    },
    {
        "question": "He not usually works on Fridays.",
        "options": [
            "He does not usually work on Fridays.",
            "He not usually works on Fridays."
        ],
        "correct": "He does not usually work on Fridays."
    },
    {
        "question": "I never have been to Australia.",
        "options": [
            "I have never been to Australia.",
            "I never have been to Australia."
        ],
        "correct": "I have never been to Australia."
    },
    {
        "question": "She always is studying for her exams.",
        "options": [
            "She is always studying for her exams.",
            "She always is studying for her exams."
        ],
        "correct": "She is always studying for her exams."
    },
    {
        "question": "Usually he wakes up early.",
        "options": [
            "He usually wakes up early.",
            "Usually he wakes up early."
        ],
        "correct": "He usually wakes up early."
    },
    {
        "question": "They sometimes are late to meetings.",
        "options": [
            "They are sometimes late to meetings.",
            "They sometimes are late to meetings."
        ],
        "correct": "They are sometimes late to meetings."
    },
    {
        "question": "Not often do we have such cold weather.",
        "options": [
            "We do not often have such cold weather.",
            "Not often do we have such cold weather."
        ],
        "correct": "We do not often have such cold weather."
    },
    {
        "question": "Never she forgets her friends’ birthdays.",
        "options": [
            "She never forgets her friends’ birthdays.",
            "Never she forgets her friends’ birthdays."
        ],
        "correct": "She never forgets her friends’ birthdays."
    },
]

# Variáveis para rastrear progresso e pontuação
current_question_index = 0
score = 0

# Widgets dinâmicos
question_label = widgets.Label()
radio_options = widgets.RadioButtons(description='Escolha a correta:', style={'description_width': 'initial'})
button = widgets.Button(description="Verificar")
output = widgets.Output()

# Ajuste de layout para espaçamento
question_label.layout = widgets.Layout(margin='10px 10px 10px 10px')
radio_options.layout = widgets.Layout(width='100%', margin='10px 10px 10px 10px')
button.layout = widgets.Layout(margin='10px 10px 10px 10px')

# Função para carregar uma questão
def load_question():
    global current_question_index
    if current_question_index < len(questions):
        question_data = questions[current_question_index]
        question_label.value = f"Pergunta {current_question_index + 1}: {question_data['question']}"
        radio_options.options = question_data["options"]
        radio_options.value = None
    else:
        # Exibir o resultado final
        with output:
            clear_output()
            print(f"Quiz concluído! Você acertou {score} de {len(questions)} questões.")
            button.disabled = True  # Desativa o botão após o final do quiz

# Função para verificar a resposta
def check_answer(b):
    global current_question_index, score
    with output:
        clear_output()
        if current_question_index < len(questions):
            selected_option = radio_options.value
            if selected_option is None:
                print("Por favor, selecione uma resposta!")
            else:
                question_data = questions[current_question_index]
                if selected_option == question_data["correct"]:
                    print("✔️ Resposta correta!")
                    score += 1
                else:
                    print(f"❌ Resposta incorreta. A resposta correta é: {question_data['correct']}")
                current_question_index += 1
                load_question()

# Associar a função ao botão
button.on_click(check_answer)

# Exibir os widgets iniciais
display(question_label, radio_options, button, output)

# Carregar a primeira questão
load_question()
```
